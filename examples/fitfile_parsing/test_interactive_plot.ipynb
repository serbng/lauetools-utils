### Let's plot just the peaks that meet a condition!
When we call the plot method of the FitFile class we just call the function ```indexation_plot```. Since this function just takes as an input a pandas DataFrame containing the data from a fit file, we can directly call it and pass a filtered dataframe according to our criteria. In this way we see which peaks in the diffraction pattern meet them

````python
from laueutils.visualization.indexation import indexation_plot
````

````python
indexation_plot(ff.peaklist.query("(h == 1 or h == 0 or h == -1) and k == 0"))
````
